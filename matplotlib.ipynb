## Keyword aguments of `add_axes()`

Resources:

1. https://matplotlib.org/stable/api/_as_gen/matplotlib.figure.Figure.add_axes.html#matplotlib.figure.Figure.add_axes

The kwargs parameter in Matplotlib's add_axes() function allows you to pass additional keyword arguments to further customize the axes. These keyword arguments can control various properties of the axes, such as labels, ticks, limits, etc. Here are some examples of how to use these keyword arguments:

Common Keyword Arguments

+ xlabel: Set the label for the x-axis.
+ ylabel: Set the label for the y-axis.
+ xlim: Set the limits for the x-axis.
+ ylim: Set the limits for the y-axis.
+ title: Set the title for the axes.
+ xticks: Set the ticks for the x-axis.
+ yticks: Set the ticks for the y-axis.

Example 1

```python
# Create a new figure
fig = plt.figure()

# Add axes with additional keyword arguments
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8],
                  xlabel='X Axis',
                  ylabel='Y Axis',
                  xlim=(0, 5),
                  ylim=(0, 10),
                  title='Custom Axes',
                  xticks=[0, 1, 2, 3, 4, 5],
                  yticks=[0, 2, 4, 6, 8, 10])

# Plot some data on the axes
ax.plot([1, 2, 3, 4], [1, 4, 9, 16])

# Display the figure
plt.show()
```

Example 2

```python
# Create a new figure
fig = plt.figure()

# Add polar axes with additional keyword arguments
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8], polar=True, title='Polar Plot')

# Generate data for a polar plot
theta = np.linspace(0, 2 * np.pi, 100)
r = np.abs(np.sin(theta))

# Plot the data on the polar axes
ax.plot(theta, r)

# Display the figure
plt.show()
```

### adjustable

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8], adjustable='datalim')
ax.plot([1, 2, 3], [1, 4, 9])
ax.set_title('adjustable="datalim"')
plt.show()
```

### anchor

Define the anchor location.

The actual drawing area (active position) of the Axes may be smaller than the Bbox (original position) when a fixed aspect is required. The anchor defines where the drawing area will be located within the available space.Either an (x, y) pair of relative coordinates (0 is left or bottom, 1 is right or top), 'C' (center), or a cardinal direction ('SW', southwest, is bottom left, etc.). str inputs are shorthands for (x, y) coordinates, as shown in the diagram in 

https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_anchor.html#matplotlib.axes.Axes.set_anchor

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8], anchor='SW')
ax.plot([1, 2, 3], [1, 4, 9])
ax.set_title('anchor="SW"') # changing it to 'N' or 'NW' or any other cardinal direction doesn't make a difference
plt.show()
```

# How to view a figure (the backend)

Resources: 

1. https://matplotlib.org/stable/users/explain/figure/backends.html#what-is-a-backend

Some common backends and associated figure views:

- Python shell -> Pop out window
- Jupyter notebook -> Inline output
- PyQT and PyGObject -> Embedded figures
- Batch scripts -> Postscript images
- Dynamic web apps -> Web images

Note: "Front-end" is just the code

There are two types of backends: user interface backends (for use in PyQt/PySide, PyGObject, Tkinter, wxPython, or macOS/Cocoa); also referred to as "interactive backends") and hardcopy backends to make image files (PNG, SVG, PDF, PS; also referred to as "non-interactive backends").
